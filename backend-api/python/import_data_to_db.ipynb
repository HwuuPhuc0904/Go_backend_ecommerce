```python
import pandas as pd
import numpy as np
import json
import ast
import re


def extract_price(price_str):
    if pd.isna(price_str):
        return 0.0
    
    price_str = str(price_str)
    match = re.search(r'[\d.,]+', price_str)
    if match:
        try:
            return float(match.group().replace(',', ''))
        except:
            return 0.0
    return 0.0

def parse_categories(cat_str):
    if pd.isna(cat_str):
        return []
    
    try:
        categories = json.loads(cat_str) if isinstance(cat_str, str) else cat_str
        return ', '.join(categories)
    except:
        try:
            categories = ast.literal_eval(cat_str) if isinstance(cat_str, str) else cat_str
            return ', '.join(categories)
        except:
            return ""

def clean_text(text):
    if pd.isna(text):
        return ""
    return str(text).replace("'", "").replace('"', "")

def parse_images(images_str):
    """Hàm chuyển đổi mảng URL hình ảnh thành chuỗi URLs phân cách bằng dấu phẩy"""
    if pd.isna(images_str):
        return ""
    
    try:
        images_array = json.loads(images_str) if isinstance(images_str, str) else images_str
        return ", ".join(images_array)
    except:
        try:
            images_array = ast.literal_eval(images_str) if isinstance(images_str, str) else images_str
            return ", ".join(images_array)
        except:
            return str(images_str)

def extract_weight(weight_str):
    if pd.isna(weight_str):
        return {'value': 0.0, 'unit': ''}
    
    weight_str = str(weight_str).strip()
    

    num_match = re.search(r'[\d.,]+', weight_str)
    
    value = 0.0
    
    if num_match:
        try:
            value = float(num_match.group().replace(',', ''))
        except:
            value = 0.0
    return value

# Lấy danh sách user_id từ bảng users (giả định đã có)
# Nếu chưa có users, hãy tạo một số users trước
try:
    users_df = pd.read_sql("SELECT id FROM users WHERE role='seller' LIMIT 100", engine)
    user_ids = users_df['id'].tolist()
    if not user_ids:
        # Nếu không có seller nào, tạo một user mặc định
        default_user_id = 1
    else:
        default_user_id = user_ids[0]
except:
    # Nếu bảng users chưa tồn tại, tạo một user mặc định
    default_user_id = 1

# Tạo DataFrame cho bảng Product
products = []

for _, row in data.iterrows():
    # Ánh xạ dữ liệu vào các trường của model Product
    product = {
        'Name': row['title'] if 'title' in data.columns and not pd.isna(row['title']) else 'Unknown Product',
        'Brand': row['brand'] if 'brand' in data.columns and not pd.isna(row['brand']) else '',
        'ProductDimension': row['product_dimensions'] if 'product_dimensions' in data.columns and not pd.isna(row['product_dimensions']) else '',
        'Description': clean_text(row['description']) if 'description' in data.columns else (
                       clean_text(row['product_description']) if 'product_description' in data.columns else ''),
        'Feature': str(row['features']) if 'features' in data.columns and not pd.isna(row['features']) else '',
        'Price': extract_price(row['final_price']) if 'final_price' in data.columns else (
                 extract_price(row['initial_price']) if 'initial_price' in data.columns else 0.0),
        'Stock': np.random.randint(10, 1000),  # Tạo số lượng tồn kho ngẫu nhiên
        'ImagesURL': parse_images(row['images']) if 'images' in data.columns and not pd.isna(row['images']) else '',
        'UserID': user_ids[np.random.randint(0, len(user_ids))] if len(user_ids) > 0 else default_user_id,
        'ParentID': row['parent_asin'] if 'parent_asin' in data.columns and not pd.isna(row['parent_asin']) else '',
        'IsAvailable': True if 'is_available' in data.columns and row['is_available'] else (
                       True if 'availability' in data.columns and not pd.isna(row['availability']) and 'in stock' in str(row['availability']).lower() else True),
        'categories': parse_categories(row['categories']) if 'categories' in data.columns else '',
        'product_weight': extract_weight(row['item_weight']) if 'item_weight' in data.columns and not pd.isna(row['item_weight']) else ''
    }
    products.append(product)

products_df = pd.DataFrame(products)

# Kiểm tra dữ liệu trước khi lưu
print("\nDữ liệu sản phẩm đã được chuẩn bị:")
print(f"Tổng số sản phẩm: {len(products_df)}")
print("\nMẫu dữ liệu:")
```

```python
try:
    # Lấy thông tin schema của bảng products
    schema_query = """
    SELECT COLUMN_NAME 
    FROM INFORMATION_SCHEMA.COLUMNS 
    WHERE TABLE_SCHEMA = DATABASE() 
    AND TABLE_NAME = 'products'
    """
    columns_in_db = pd.read_sql(schema_query, engine)
    print("Các cột có trong bảng products:")
    db_columns_list = columns_in_db['COLUMN_NAME'].tolist()
    print(db_columns_list)
    
    # In ra tên các cột trong DataFrame để so sánh
    print("\nCác cột trong DataFrame:")
    df_columns_list = products_df.columns.tolist()
    print(df_columns_list)
    
    # Chuyển đổi tất cả tên cột sang chữ thường để so sánh không phân biệt chữ hoa/thường
    df_columns_lower = [col.lower() for col in df_columns_list]
    db_columns_lower = [col.lower() for col in db_columns_list]
    
    # Tìm các cột không khớp
    missing_in_db = [col for col in df_columns_lower if col not in db_columns_lower]
    print("\nCác cột trong DataFrame không có trong bảng:")
    print(missing_in_db)
    
    # Mapping tên cột - điều chỉnh cho chính xác
    column_mapping = {
        'productdimension': 'product_dimension', 
        'imageurl': 'images_url',  # Kiểm tra tên chính xác, có thể là 'imageurl' không phải 'imagesurl'
        'imagesurl': 'images_url',
        'image_url': 'images_url',  # Thêm trường hợp khác
        'userid': 'user_id',
        'parentid': 'parent_id',
        'isavailable': 'is_available',
    }
    
    # Tạo mapping từ tên cột trong DataFrame sang tên cột trong DB
    rename_dict = {}
    for df_col in df_columns_list:
        df_col_lower = df_col.lower()
        if df_col_lower in column_mapping:
            rename_dict[df_col] = column_mapping[df_col_lower]
    
    # Điều chỉnh tên cột trong DataFrame
    products_df = products_df.rename(columns=rename_dict)
    
    print("\nTên cột mới trong DataFrame sau khi mapping:")
    print(products_df.columns.tolist())
    
    # Tạo một từ điển ánh xạ tên cột DB (chữ thường) đến tên cột thực tế
    db_column_case_map = {col.lower(): col for col in db_columns_list}
    
    # Lọc và giữ lại các cột có trong bảng DB và đảm bảo đúng chữ hoa/thường
    products_df_filtered = pd.DataFrame()
    for col in products_df.columns:
        col_lower = col.lower()
        if col_lower in db_columns_lower:
            # Sử dụng tên cột chính xác như trong DB
            db_col_name = db_column_case_map[col_lower]
            products_df_filtered[db_col_name] = products_df[col]
    
    print(f"\nSố lượng cột sau khi lọc: {len(products_df_filtered.columns)}")
    print("Các cột được giữ lại:")
    print(products_df_filtered.columns.tolist())
    
    # Kiểm tra xem các cột bạn quan tâm có trong DataFrame không
    important_columns = ['product_dimension', 'images_url', 'parent_id', 'is_available']
    for col in important_columns:
        if col in products_df_filtered.columns:
            print(f"Cột {col} có trong dữ liệu sẽ chèn")
            # Hiển thị một vài giá trị mẫu
            print(f"Mẫu dữ liệu: {products_df_filtered[col].head(3).tolist()}")
        else:
            print(f"Cột {col} KHÔNG có trong dữ liệu sẽ chèn")
    
    # Lưu dữ liệu đã điều chỉnh vào MySQL
    print("\nĐang lưu dữ liệu vào MySQL...")
    products_df_filtered.to_sql('products', engine, if_exists='append', index=False)
    
    print("Đã nhập thành công dữ liệu vào bảng products!")
    
    # Kiểm tra số lượng bản ghi đã nhập
    count_query = "SELECT COUNT(*) as count FROM products"
    result = pd.read_sql(count_query, engine)
    print(f"Tổng số bản ghi trong bảng products: {result['count'].iloc[0]}")
    
except Exception as e:
    print(f"Lỗi: {str(e)}")
    # In thêm thông tin chi tiết lỗi
    import traceback
    traceback.print_exc()
```
